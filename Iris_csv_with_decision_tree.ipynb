# data

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/IRIS.csv')
```

```python
df
```

> **4 features**

```python
print(f'# of rows before drop : {len(df)}')

df.dropna(inplace=True)

print(f'# of rows after drop : {len(df)}')
```

```python
from sklearn import tree
y = df.pop('species') 
X = df

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
clf.score(X,y)
```

```python
clf.predict(X=[[5.1,3.5,1.4,0.2 ]])
```

```python
import pickle

pickle.dump(clf, open('model.pkl','wb'))
```
